```python
import numpy as np
import matplotlib.pyplot as plt
```

Define angles to be used for the stability plots.

```python
# Number of points for array initialization
N = 20
# Generate radian angles
angles = np.linspace(0 + 2*np.pi/N, 2*np.pi, N)
# Generate linear domain of w*dt values
wdts = np.linspace(0, 3, N)
```

### Problem 2, part c - stability region, part (a)

```python
poly_a = [np.array([(1j), 1-np.exp((1j)*theta)]) for theta in angles]
roots_a = [np.roots(poly_a[i]) for i in range(0, len(poly_a))]
```

### Problem 2, part c - stability region, part (b)

```python
# Calculate values of alpha (polynomial after plugging alpha for (b) is complicated)
a_vals = np.linspace(0, 10, 1)
# Loop over values of alpha and generate the corresponding complex plane values
poly_b = [np.array([a**2/4, 0, -(a-1/4), 1-np.exp(1j*theta)]) for a in a_vals for theta in angles ]
roots_b = [np.roots(poly_b[i]) for i in range(0, len(poly_b))]
```

```python
# Calculate amplification factor magnitudes
mags_a = [np.sqrt(1+x**2) for x in wdts]
mags_b = [np.sqrt(1 - (1/4 - (-np.sqrt(x**2+4)-2)/x**2)*x**4 + (1/4)*((-np.sqrt(x**2+4)-2)/x**2)**2*x**6) for x in wdts]
```

```python
# Calculate relative phase error magnitudes
rpe_a = [np.arctan(x)/x for x in wdts]
rpe_b = [ for x in wdts]
```

### Amplification factor plots

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(wdts, mags_a, label='Part (a)')
ax.plot(wdts, mags_b, label='Part (b)')

# Plot metadata
ax.set_aspect('equal')
ax.grid()
ax.set_xlabel('$\omega \Delta t$', labelpad=15)
ax.set_ylabel('|A|', labelpad=15)
ax.set_ylim([0, 3])

fig.tight_layout()

x_pos = 0.55
fig.suptitle('Magnitude of $A$', x=x_pos, y=1.05)
fig.legend(ncol=2, frameon=False, loc='upper center', bbox_to_anchor=(x_pos, 1));
```

### Relative phase error plots

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(wdts, rpe_a, label='Part (a)')
ax.plot(wdts, rpe_b, label='Part (b)')

# Plot metadata
ax.set_aspect('equal')
ax.grid()
ax.set_xlabel('$\omega \Delta t$', labelpad=15)
ax.set_ylabel('R', labelpad=15)

x_pos = 0.55
fig.suptitle('Relative phase errors', x=x_pos, y=1.05)
fig.legend(ncol=2, frameon=False, loc='upper center', bbox_to_anchor=(x_pos, 1))

fig.tight_layout()
```

### Stability region plots

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(np.real(roots_a), np.imag(roots_a), marker='o', label='Part (a)')
ax.plot(np.real(roots_b), np.imag(roots_b), marker='o', label='Part (b)')

# Plot metadata
ax.set_aspect('equal')
ax.grid()
ax.set_xlabel('Re($\omega \Delta t$)', labelpad=15)
ax.set_ylabel('Im($\omega \Delta t$)', labelpad=15)

x_pos = 0.55
fig.suptitle('Magnitude of $A$', x=x_pos, y=1.05)
fig.legend(ncol=2, frameon=False, loc='upper center', bbox_to_anchor=(x_pos, 1));
fig.tight_layout()
```

### Problem 3: Lorenz equation discretization

```python
# Define functions
def F_x(x, y, z, t, r):
    return -3*(x - y)
def F_y(x, y, z, t, r):
    return -x*z + r*x - y
def F_z(x, y, z, t, r):
    return x*y - z

def diff_x(x, y, z, t, dt, r):
    a, b, c = 1, 1/2, 1
    xi_1 = x
    xi_2 = x + dt*a*F_x(x, y, z, t, r)
    x_ = x + dt*(b*F_x(xi_1, y, z, t + c*dt, r) + b*F_x(xi_2, y, z, t + c*dt, r))
    return x_
def diff_y(x, y, z, t, dt, r):
    a, b, c = 1, 1/2, 1
    xi_1 = y
    xi_2 = y + dt*a*F_y(x, y, z, t, r)
    y_ = y + dt*(b*F_y(x, xi_1, z, t + c*dt, r) + b*F_y(x, xi_2, z, t + c*dt, r))
    return y_
def diff_z(x, y, z, t, dt, r):
    a, b, c = 1, 1/2, 1
    xi_1 = z
    xi_2 = z + dt*a*F_z(x, y, z, t, r)
    z_ = z + dt*(b*F_z(x, y, xi_1, t + c*dt, r) + b*F_z(x, y, xi_2, t + c*dt, r))
    return z_

# Define initial values
r = 16
x_0, y_0, z_0 = 1, 1, 1
x = np.array([x_0])
y = np.array([y_0])
z = np.array([z_0])

# Define time and timestep
dt, t_0 = 0.01, 0
t = np.array([t_0])
N = 15000 # number of steps

for i in range(0, N):
    #print('Time step: {0} | Time: {1}'.format(i, t[i]))
    
    x_ = diff_x(x[i], y[i], z[i], t[i], dt, r)
    x = np.append(x, x_)
    
    y_ = diff_y(x[i], y[i], z[i], t[i], dt, r)
    y = np.append(y, y_)
    
    z_ = diff_z(x[i], y[i], z[i], t[i], dt, r)
    z = np.append(z, z_)
    
    # Step forward in time
    t = np.append(t, t[i] + dt)
```

```python
fig, ax = plt.subplots(ncols=2, sharey=True)
ax[0].plot(x, y)
ax[1].plot(x, z)
```
